```python
from __future__ import print_function
__author__= 'fbb'
import os
import numpy as np
from scipy.optimize import curve_fit, minimize
import pylab as plt
from scipy.stats import norm
%pylab inline
```

```python
#distributions and universal variables setup

md = ['chisq', 'pois', 'wald', 'binomial', 'gamma', 'normal']
test_mean = 100 #mean
md = {} #md is an empty dictionary
means = {} #means is an empty dictionary
mysize = 2000/(np.array(range(1,100)+[10]))
allmeans= []
```

```python
#Chi-squared
mysize=2000/(np.array(range(1,100)+[10]))

md['chisq']={} 

for n in mysize:
    md['chisq'][n] = np.random.chisquare(test_mean, size=n)
    
means = {}
means['chisq'] = {}
fig_mu_N = plt.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['chisq'].iterkeys():
    means['chisq'][nn] = md['chisq'][nn].mean()
    axchisq_mu_n.plot(nn, means['chisq'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysize), max(mysize)],[test_mean,test_mean], 'k')

for n in means['chisq']:
    allmeans.append(means['chisq'][n])
```

```python
#Poisson distribution

md['pois']={} 

for n in mysize:
    md['pois'][n] = np.random.poisson(test_mean, size=n)
    
means = {}
means['pois'] = {}
fig_mu_N = plt.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['pois'].iterkeys():
    means['pois'][nn] = md['pois'][nn].mean()
    axchisq_mu_n.plot(nn, means['pois'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysize), max(mysize)],[test_mean,test_mean], 'k')

for n in means['pois']:
    allmeans.append(means['pois'][n])
```

```python
#Gamma distribution

md['gamma']={} 

for n in mysize:
    md['gamma'][n] = np.random.gamma(test_mean, scale = 1, size=n)
    
means = {}
means['gamma'] = {}
fig_mu_N = plt.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['gamma'].iterkeys():
    means['gamma'][nn] = md['gamma'][nn].mean()
    axchisq_mu_n.plot(nn, means['gamma'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysize), max(mysize)],[test_mean,test_mean], 'k')

for n in means['gamma']:
    allmeans.append(means['gamma'][n])
```

```python
#Binomial distribution

md['binomial']={} 

for x in mysize:
    md['binomial'][x] = np.random.binomial(n = 200, p = 0.5, size=x)
    
means = {}
means['binomial'] = {}
fig_mu_N = plt.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['binomial'].iterkeys():
    means['binomial'][nn] = md['binomial'][nn].mean()
    axchisq_mu_n.plot(nn, means['binomial'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysize), max(mysize)],[test_mean,test_mean], 'k')

for n in means['binomial']:
    allmeans.append(means['binomial'][n])
```

```python
#Lognormal distribution

md['lnorm']={} 

for n in mysize:
    md['lnorm'][n] = np.random.lognormal(mean = 4.1, sigma = 1, size=n)
    
means = {}
means['lnorm'] = {}
fig_mu_N = plt.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['lnorm'].iterkeys():
    means['lnorm'][nn] = md['lnorm'][nn].mean()
    axchisq_mu_n.plot(nn, means['lnorm'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysize), max(mysize)],[test_mean,test_mean], 'k')

for n in means['lnorm']:
    allmeans.append(means['lnorm'][n])
```

```python
#Gaussian distribution

md['gaus']={} 

for n in mysize:
    md['gaus'][n] = np.random.normal(test_mean,1, size=n)
    
means = {}
means['gaus'] = {}
fig_mu_N = plt.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['gaus'].iterkeys():
    means['gaus'][nn] = md['gaus'][nn].mean()
    axchisq_mu_n.plot(nn, means['gaus'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysize), max(mysize)],[test_mean,test_mean], 'k')

for n in means['gaus']:
    allmeans.append(means['gaus'][n])
```

In all of these distributions we see an approximately normal curve develop with a mean of around 100. This makes sense for all of the distributions, since we started off with a population mean of around 100, and the Central Limit Theorem tells us that an increasing number of sample means taken off of any distribution will eventually start to approximate a normal curve.

```python
plt.figure(figsize=(10,10))
plt.hist(allmeans,bins=30)
plt.xlabel('sample mean')
plt.ylabel('N')
```
