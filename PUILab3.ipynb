```python
#from __future__ import print_function
import pylab as pl
import numpy as np
%pylab inline
```

```python
mymean = 100
mydistribution = np.random.chisquare(mymean, 100)
pl.hist(mydistribution)
```

```python
#sizes = range(10, 2000, 100)
sizes = 2000/(np.array(range(1,90))+10)
print sizes
```

```python
distributions = {}
distributions['chisquare'] = {}
```

```python
for i in sizes:
    distributions['chisquare'][i] = np.random.chisquare(mymean, i)
```

```python
fig_mu_N = pl.figure(figsize = (10, 10))
axchisq_mu_n = fig_mu_N.add_subplot(111)

for n in distributions['chisquare'].iterkeys():
    axchisq_mu_n.plot(n, distributions['chisquare'][n].mean(), 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(sizes), max(sizes)], [mymean,mymean], 'k')
#print distributions['chisquare']
```
