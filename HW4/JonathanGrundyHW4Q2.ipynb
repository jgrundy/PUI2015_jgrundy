```python
import pylab as pl
import pandas as pd
import numpy as np
import os
import scipy.stats as sps
%pylab inline
```

```python
#generate the distribution
dist_n = np.random.randn(10000)
dist_p = np.random.poisson(1, 10000)
```

```python
#test ad and ks
print "normal on normal", sps.kstest(dist_n,'norm')
print "normal on normal", sps.anderson(dist_n, dist='norm')
print "" 

print "poisson on normal", sps.kstest(dist_p,'norm')
print "poisson on normal", sps.anderson(dist_p, dist='norm')

threshold = sps.anderson(dist_n, dist='norm')[1][sps.anderson(dist_n, dist='norm')[2]==[1.0]]
print threshold
```

```python
#BINOMIAL

#to store the data i am generating an empty array
narray = range(1,50,1)
ks_b = np.zeros(len(narray))
ad_b = np.zeros(len(narray))
kl_b = np.zeros(len(narray))

def mynorm (x, mu, var):
    return sps.norm.cdf(x, loc=mu, scale=var)

#CHANGED THE P VALUE
p=0.99

for i,n in enumerate(narray):
    p=0.9 #parameter for the binomial, my arbitrary choice
    #generate the distribution
    dist = np.random.binomial(n, p, 10000)
    #run the tests. 
 
    ks_b[i] = sps.kstest(dist, mynorm, args=(n*p, n*p*(1.0-p)))[0]
    ad_b[i] = sps.anderson(dist, dist='norm')[0]
    
     # for KL and Pearson's chisq I have to simulate the normal distribution as well
    mybins=np.linspace(min(dist),max(dist), 10) 
    bincenters = mybins[:-1]+0.5*(mybins[1]-mybins[0])

    kl_b [i] =  sps.entropy(np.histogram(dist, bins=mybins)[0], sps.norm.pdf(bincenters, loc=n*p, scale=n*p*(1.0-p)))
```

```python
fig = pl.figure(figsize = (15,5))
fig.add_subplot(131)
pl.plot(narray, ks_b, label='KS')
pl.legend()

fig.add_subplot(132)
pl.plot(narray, ad_b,  label='AD')
pl.plot([narray[0], narray[-1]],[threshold, threshold])
pl.plot()
pl.plot()
pl.legend()

fig.add_subplot(133)
pl.plot(narray, kl_b, label='K-L ')

pl.legend()
```

```python
#POISSON 

#to store the data i am generating an empty array of the size of the values of n i want to test. i do that with np.zeros which takes the length of the array as argument, and the data type as optional argument, but default is float, which is fine by me, so i do not need any argument other then the lenght 
narray = range(1,50,1)
ks_b = np.zeros(len(narray))
ad_b = np.zeros(len(narray))
kl_b = np.zeros(len(narray))

for i,n in enumerate(narray):
    p=0.8 #parameter for the binomial, my arbitrary choice
    #generate the distribution
    dist = np.random.poisson(n, 10000)
    #run the tests. 
    ks_b[i] = sps.kstest(dist, mynorm, args=(n, sqrt(n)))[0]
    ad_b[i] = sps.anderson(dist, dist='norm')[0]
     
    # for KL and Pearson's chisq I have to simulate the normal distribution as well
    mybins = np.linspace(min(dist),max(dist), 10) 
    bincenters = mybins[:-1]+0.5*(mybins[1]-mybins[0])
 
    kl_b [i] =  sps.entropy(np.histogram(dist, bins=mybins)[0], sps.norm.pdf(bincenters, loc=n*p, scale=n*p*(1.0-p)))
```

```python

fig = pl.figure(figsize = (15,5))
fig.add_subplot(131)
pl.plot(narray, ks_b, label='KS')
pl.legend()

fig.add_subplot(132)
pl.plot(narray, ad_b,  label='AD')
pl.plot([narray[0], narray[-1]],[threshold, threshold])
pl.plot()
pl.plot()
pl.legend()

fig.add_subplot(133)
pl.plot(narray, kl_b, label='K-L ')

pl.legend()
```

For both sets of graphs, increasing the p or n values helps to smooth these graphs. What that means is that as we increase our p value or increase our sample size (n), the samples that are being taken more closely represent or match the curves that they're supposed to be approximating.
